```python
max_flex = 100
flexibility = range(0,max_flex)
```

```python
import math 

bias = []
for i in flexibility:
    bias.append((max_flex-i)**2)
```

```python
variance = []
for i in flexibility:
    variance.append((i)**2)
```

```python
bayes_error = [4500]*max_flex
```

```python
train_error = []
variance_rate = []

for i in flexibility:
    variance_rate.append((max_flex-i)/max_flex)
    
for i in flexibility:
    train_error.append((bias[i] + (variance[i]*variance_rate[i]) + bayes_error[i]*0.1))
```

```python
test_error = []
for i in flexibility:
    test_error.append(bias[i] + variance[i] + bayes_error[i]*0.2)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
plt.plot(flexibility, bias, label = 'bias')
plt.plot(flexibility, variance, label = 'variance')
plt.plot(flexibility, bayes_error, label = 'irreducible error')
plt.plot(flexibility, train_error, label = 'train error')
plt.plot(flexibility, test_error, label = 'test error')
plt.xlabel("Flexibility")
plt.legend()
```
